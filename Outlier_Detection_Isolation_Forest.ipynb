# Isolation Forest
It is an unsupervised learning approach in detecting outliers. 

```python
import pandas as pd
import numpy as np
from sklearn import base
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import datasets
from sklearn.ensemble import IsolationForest
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt
```

```python
x_train = datasets.load_iris().data[:,:2]
x_test = np.random.random((15,2))*np.array([7,3])+np.array([2,1])
```

```python
clf = IsolationForest(max_samples=100, random_state=0)
clf.fit(x_train)
```

```python
def plot(X, c='blue'):
    plt.scatter(X[:,0], X[:,1], s=30, c=c)
    
plot(x_train, c='blue')
plot(x_test, c="red")

xx, yy = np.meshgrid(np.linspace(2.5, 9, 500), np.linspace(0.5, 5, 500))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), 0, 7), cmap=plt.cm.Blues_r)
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='black')

plot(x_train, c='blue')
plot(x_test,c="red")
```
